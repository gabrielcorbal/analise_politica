```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import glob
import seaborn as sns
pd.set_option('display.max_columns', 500)
plt.rc('figure', figsize=(10, 5))
```

# 1 Turno

```python
path = r'C:\Users\Gabriel\Desktop\analise_politica\1turno' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

df1turno = pd.concat(li, axis=0, ignore_index=True)
```

```python
df1turno
```

```python
dfnaovotaram1 = df1turno[df1turno.SG_PARTIDO == 'Não Votaram']
```

```python
dfnaovotaram1
```

```python
sns.set(style="whitegrid")

ax = sns.barplot(x="Ano", y="Porcentagem", data=dfnaovotaram1, color="salmon")
```
